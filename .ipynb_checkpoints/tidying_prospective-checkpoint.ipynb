```python
import pandas as pd

data = pd.read_csv('2025_07_16.csv', header = None)
data.columns = ['filename', 'probability']
data
```

```python
def spliting(df):
    filename = df['filename'].str.split('.').apply(lambda x: x[0])
    df['NHI'] = filename.str.split('_').apply(lambda x: x[0])
    df['eye'] = filename.str.split('_').apply(lambda x: x[1])
    df['slice'] = filename.str.split('_').apply(lambda x: x[2]).astype('int')
    df = df[['NHI', 'eye', 'slice', 'probability']]
    return df
```

```python
data = spliting(data)
```

```python
import datetime

data['date'] = pd.to_datetime(datetime.date(2025, 7, 16))
data
```

```python
data.to_csv('2025_07_16_results.csv')
```

```python
import pandas as pd

normal_df = pd.read_csv("normal.csv", header = None)
normal_df.columns = ["filename", "probability"]
```

```python
normal_df['NHI'] = normal_df['filename'].str.split('.').apply(lambda x: x[0]).str.split('_').apply(lambda x: x[0])
normal_df['eye'] = normal_df['filename'].str.split('.').apply(lambda x: x[0]).str.split('_').apply(lambda x: x[1])
normal_df['slice'] = normal_df['filename'].str.split('.').apply(lambda x: x[0]).str.split('_').apply(lambda x: x[2]).astype('int')
normal_df['HCQ_toxicity'] = 0
```

```python
normal_df
```

```python
toxicity_df = pd.read_csv("toxicity.csv", header = None)
toxicity_df.columns = ["filename", "probability"]
toxicity_df['NHI'] = toxicity_df['filename'].str.split('.').apply(lambda x: x[0]).str.split('_').apply(lambda x: x[0])
toxicity_df['eye'] = toxicity_df['filename'].str.split('.').apply(lambda x: x[0]).str.split('_').apply(lambda x: x[1])
toxicity_df['slice'] = toxicity_df['filename'].str.split('.').apply(lambda x: x[0]).str.split('_').apply(lambda x: x[2]).astype('int')
toxicity_df['HCQ_toxicity'] = 1
toxicity_df
```

```python
combined = pd.concat([normal_df, toxicity_df])
combined = combined[['NHI', 'eye', 'slice', 'probability', 'HCQ_toxicity']]
combined
```

```python
combined.to_csv('results.csv')
```

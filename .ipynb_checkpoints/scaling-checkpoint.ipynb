```python
import pandas as pd

data = pd.read_csv('results.csv')
data = data[data.columns[1:]]
data
```

```python
data['normalised_slice'] = data.groupby(['NHI', 'eye'])['slice'].transform(lambda x: (x - x.mean()) / (x.max() - x.min()))
```

```python
import seaborn as sns
sns.lmplot(data = data,
            x = 'normalised_slice',
            y = 'probability',
            hue = 'HCQ_toxicity',
            lowess = True,
            scatter_kws = {'alpha':0.4, 's':0.5} # s to change size of markers
            )
```

```python
sns.lmplot(data = data,
            x = 'normalised_slice',
            y = 'probability',
            col = 'HCQ_toxicity',
            hue = 'HCQ_toxicity',
            lowess = True,
            scatter_kws = {'alpha':0.4, 's':0.5} # s to change size of markers
            )
```

```python
#export cases classified as normal but deemed to have high probability of HCQ in the central slices
data[(data['normalised_slice'] < 0.2) & (data['normalised_slice'] > -0.2) & (data['HCQ_toxicity'] == 0) & (data['probability'] >= 0.5)].to_csv('high_probability_normals.csv')
```
